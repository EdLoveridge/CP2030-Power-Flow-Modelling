```python
# measure distances between all of the node geometries in km
import ast  # to safely parse the stringified dict in 'geo'

# Example: your bus dataframe (replace with your actual df)
# df = pd.read_csv("your_bus_file.csv")  # if loading from file

# Parse the 'geo' column into actual coordinate lists
net.bus["coords"] = net.bus["geo"].apply(lambda g: ast.literal_eval(g)["coordinates"])

# Convert to separate x, y columns
net.bus["x"] = net.bus["coords"].apply(lambda c: c[0])
net.bus["y"] = net.bus["coords"].apply(lambda c: c[1])

# Function to compute distance between two buses by name
def bus_distance(net, bus1, bus2):
    x1, y1 = net.bus.loc[net.bus["name"] == bus1, ["x", "y"]].values[0]
    x2, y2 = net.bus.loc[net.bus["name"] == bus2, ["x", "y"]].values[0]
    return np.sqrt((x2 - x1)**2 + (y2 - y1)**2)

# Distances between specific bus pairs

AB =  bus_distance(net, "A", "B")/1000
AC =  bus_distance(net, "A", "C")/1000
QS =  bus_distance(net, "Q", "S")/1000
PN =  bus_distance(net, "P", "N")/1000
LN =  bus_distance(net, "L", "N")/1000
ST =  bus_distance(net, "S", "T")/1000
RS =  bus_distance(net, "R", "S")/1000
NR =  bus_distance(net, "N", "R")/1000
MN =  bus_distance(net, "M", "N")/1000
LM =  bus_distance(net, "L", "M")/1000
KP =  bus_distance(net, "K", "P")/1000
PQ =  bus_distance(net, "P", "Q")/1000
KL =  bus_distance(net, "K", "L")/1000
JK =  bus_distance(net, "J", "K")/1000
DJ =  bus_distance(net, "D", "J")/1000
GL =  bus_distance(net, "G", "L")/1000
GH =  bus_distance(net, "G", "H")/1000
AD =  bus_distance(net, "A", "D")/1000
GB =  bus_distance(net, "G", "B")/1000
EG =  bus_distance(net, "E", "G")/1000
EF =  bus_distance(net, "E", "F")/1000
DL =  bus_distance(net, "D", "L")/1000
JC =  bus_distance(net, "J", "C")/1000
KN = bus_distance(net, "K", "N")/1000


TK = bus_distance(net, "T", "K")/1000
SH = bus_distance(net, "S", "H")/1000

print("Distance T-K (km):", TK)
print("Distance S-H (km):", SH)
```

```python
import ast
import geopandas as gpd
from shapely.geometry import shape, Point, LineString

def build_geodata_from_net(net, epsg_guess_for_projected=27700):
    # --- helper to parse the 'geo' cell which may be a dict-like object or a string ---
    def parse_geo_cell(cell):
        if cell is None:
            return None
        # if already a dict-like mapping with 'coordinates' -> use shape
        if isinstance(cell, dict):
            return shape(cell)
        # if it's a shapely geometry already
        try:
            from shapely.geometry.base import BaseGeometry
            if isinstance(cell, BaseGeometry):
                return cell
        except Exception:
            pass
        # if it's a string representation like '{"coordinates":[...], "type":"Point"}'
        if isinstance(cell, str):
            try:
                obj = ast.literal_eval(cell)
                return shape(obj)
            except Exception:
                # try json-like parsing fallback
                try:
                    import json
                    obj = json.loads(cell)
                    return shape(obj)
                except Exception:
                    raise ValueError(f"Cannot parse geo column value: {cell!r}")
        raise ValueError(f"Unsupported geo cell type: {type(cell)}")
    
    # --- Build bus_geodata if missing ---
    if (not hasattr(net, "bus_geodata")) or (net.bus_geodata is None):
        bus_df = net.bus.copy()
        # require 'geo' column
        if "geo" not in bus_df.columns:
            raise ValueError("net.bus has no 'geo' column — can't build bus_geodata automatically.")
        geoms = []
        for i, val in bus_df["geo"].items():
            if val is None:
                geoms.append(None)
            else:
                geoms.append(parse_geo_cell(val))
        bus_gdf = gpd.GeoDataFrame(bus_df, geometry=geoms)
        # detect whether coordinates are lon/lat or projected by inspecting first valid point
        sample = next((p for p in bus_gdf.geometry if p is not None), None)
        if sample is None:
            raise RuntimeError("No geometries found in net.bus.geo")
        x, y = sample.x, sample.y
        # if coordinates look like lon/lat (within -180..180, -90..90) assume EPSG:4326
        if abs(x) <= 180 and abs(y) <= 90:
            inferred_epsg = 4326
        else:
            # otherwise assume projected (for UK likely EPSG:27700). You can change epsg_guess_for_projected.
            inferred_epsg = epsg_guess_for_projected
        bus_gdf.set_crs(epsg=inferred_epsg, inplace=True)
        net.bus_geodata = bus_gdf

    # --- Build line_geodata if missing ---
    if (not hasattr(net, "line_geodata")) or (net.line_geodata is None):
        if not hasattr(net, "bus_geodata") or net.bus_geodata is None:
            raise RuntimeError("bus_geodata is required to build line_geodata")
        # map bus index -> point geometry
        bus_geom = net.bus_geodata.geometry.to_dict()
        line_geoms = []
        for idx, row in net.line.iterrows():
            fb = row["from_bus"]
            tb = row["to_bus"]
            p1 = bus_geom.get(fb)
            p2 = bus_geom.get(tb)
            if (p1 is None) or (p2 is None):
                line_geoms.append(None)
            else:
                line_geoms.append(LineString([p1, p2]))
        line_gdf = gpd.GeoDataFrame(net.line.copy(), geometry=line_geoms)
        # set same CRS as buses
        line_gdf.set_crs(net.bus_geodata.crs, inplace=True)
        net.line_geodata = line_gdf

    return net

# --- Usage: build geodata and reproject to WebMercator (EPSG:3857) for basemap plotting ---
net = build_geodata_from_net(net, epsg_guess_for_projected=27700)

# optional: check inferred CRS
print("bus_geodata CRS:", net.bus_geodata.crs)
print("sample bus coords (first):", net.bus_geodata.geometry.iloc[0].x, net.bus_geodata.geometry.iloc[0].y)

# reproject to EPSG:3857 *in place* so plotting with contextily works
net.bus_geodata = net.bus_geodata.to_crs(epsg=3857)
net.line_geodata = net.line_geodata.to_crs(epsg=3857)

# copy results into the geodataframe so geopandas plotting can colour the lines
net.line_geodata["loading_percent"] = net.res_line["loading_percent"]

print("Reprojected to:", net.bus_geodata.crs)
```
